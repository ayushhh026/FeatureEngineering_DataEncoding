```python
#Data Encoding 
#Model only understands numerical data so we have to convert categorial data to numerical data we do this by data encoding
 
```

```python
#1. Nominal/One Hot Encoding (OHE)
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder
```

```python
##Creating a simple DF
df = pd.DataFrame({
    'color': ['red','blue','green','green','red']
})
```

```python
df
```

```python
## Creating instance of One hot Encoder
encoder=OneHotEncoder()
```

```python
## Perform Fit (learn about all the categories) and Transform (convert the learned information to data 0,1)
#  either use fit() and transform() seperately or in one line
encoded=encoder.fit_transform(df[['color']]).toarray()# does alphabet sorting starts with Blue then Green then Red
```

```python
encoded
```

```python
import pandas as pd
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
## For new data to check 
encoder.transform([['blue'],['red']]).toarray()
```

```python
pd.concat([df,encoder_df],axis=1)
```
