```python
#2.Label Encoding and Ordinal Encoding
# assigns unique numerical to each catergory variable
# Red : 1 Green : 2 Blue : 3
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame({
    'color': ['red','blue','green','green','red']
})
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder=LabelEncoder()
```

```python
lbl=lbl_encoder.fit_transform(df[['color']])
```

```python
lbl_encoder.transform([['red']])
```

```python
lbl_df=pd.DataFrame(lbl,columns=['lbl'])
```

```python
pd.concat([df,lbl_df],axis=1)
```
