```python
##4.Target Guided Oridnal Encoding
# Technique used to encode cat variables based on their relationship with the target variable

```

```python
import pandas as pd 

df= pd.DataFrame({
    'city': ['New York', 'London', 'Paris', 'Tokyo', 'New York', 'Paris'],
    'price': [200,150,300,250,180,320]#assume this is target variable
})
```

```python
df
```

```python
mean_price=df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoded']=df['city'].map(mean_price)
```

```python
df
```

```python
df[['city_encoded','price']]
```

```python
df_label=df.groupby('city')['price'].mean().sort_values().index
df_mapping={k:i for i,k in enumerate(df_label)}
```

```python
df_mapping
```

```python
df['updated']=df['city'].map(df_mapping)
```

```python
df[['city','updated','price']]
```

```python
#Time based on total_bill target 
import seaborn as sns
df = sns.load_dataset('tips')
```

```python
df
```

```python
df['time'].value_counts()
```

```python
df['total_bill']
```

```python
mean_price=df.groupby('time')['total_bill'].mean().to_dict()
```

```python
mean_price
```

```python
df['time_encoded']=df['time'].map(mean_price)
```

```python
df[['time','time_encoded','total_bill']]
```

```python
label=df.groupby('time')['total_bill'].mean().sort_values().index
df_mapping={k:i for i,k in enumerate(label)}
```

```python
df_mapping
```

```python
df['updated']=df['time'].map(df_mapping)
```
