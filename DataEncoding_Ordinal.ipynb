```python
#3.Ordinal encoding :- used to encode categorial data that have an ordeer or ranking
```

```python
import pandas as pd 
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.DataFrame({
    'size': ['small','medium','large','medium','small','large']
})
```

```python
df
```

```python
# create Instance 
encoder=OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoded=encoder.fit_transform(df[['size']])
```

```python
encoded
```

```python
encoder.transform([['small']])
```

```python
od=pd.DataFrame(encoded,columns=['Rankings'])
pd.concat([df,od],axis=1)
```
